```
import math

def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    else:
      sqrt_n = int(math.sqrt(n))
      for i in range(5, sqrt_n + 1, 6):
          if n % i == 0 or n % (i + 2) == 0:
              return False
      return True
Nmax = 10000
domain = [n for n in range(6,Nmax) if not is_prime(n)]
print(len(domain))

domain = cp_model.Domain.FromValues(domain)
domain
```

```
def ff(N=3):
  model = cp_model.CpModel()
  solver = cp_model.CpSolver()
  nodes = range(N)


  U = {i:model.NewIntVarFromDomain(domain,f"connection_{i}") for i in nodes}


  for i in nodes:
    if i+1 in nodes:
      model.Add(U[i+1]-U[i] <= N-1)
      model.Add(U[i+1]-U[i] >= 1)

  model.Add(U[N-1]-U[0] <= (N-1)**2)

  expressions = [U[n] for n in nodes]
  model.AddAllDifferent(expressions)

  model.Add(U[0]==6)

  model.Minimize(sum(expressions))

  solver.parameters.max_time_in_seconds = 600
  status = solver.Solve(model)

  """
  if status == cp_model.OPTIMAL:
    print("OPTIMAL")
  elif status == cp_model.FEASIBLE:
    print("FEASIBLE")
  elif status == cp_model.INFEASIBLE:
    print("INFEASIBLE")
  elif status == cp_model.UNKNOWN:
    print("UNKNOWN")
  """
  #print(solver.ObjectiveValue())
  #print(f"x", solver.Value(x))
  a = [solver.Value(U[n]) for n in nodes]
  print(a)
  return a
```
